```python
import os, json, time, random
from pathlib import Path
from dataclasses import dataclass
from typing import Dict, List, Tuple

import numpy as np
import pandas as pd

import torch
import torch.nn as nn
import torch.nn.functional as F
from torch.utils.data import Dataset, DataLoader
from torchvision import transforms
from torchvision.models import googlenet, GoogLeNet_Weights
from PIL import Image

import timm

from sklearn.model_selection import train_test_split
from sklearn.metrics import (
    accuracy_score, precision_recall_fscore_support, confusion_matrix,
    roc_auc_score, roc_curve
)
from sklearn.preprocessing import label_binarize

import matplotlib.pyplot as plt
from fvcore.nn import FlopCountAnalysis


# =============================
# Config (YOUR REQUEST)
# =============================
@dataclass
class CFG:
    DATA_DIR: str = "/kaggle/input/datasets/fernando2rad/brain-tumor-mri-images-44c"
    WORK_DIR: str = "/kaggle/working/"
    seed: int = 42

    img_size: int = 224
    batch_size: int = 64
    num_workers: int = 2

    # exactly one split
    train_pct: int = 90
    test_pct: int = 10
    val_in_train: float = 0.10   # 10% of train used for validation

    # exactly 50 epochs, fully done
    epochs: int = 50

    # patience asked (kept, but not used to stop early since we force full 50)
    patience: int = 7

    lr: float = 3e-4
    weight_decay: float = 1e-4

cfg = CFG()
os.makedirs(cfg.WORK_DIR, exist_ok=True)

DEVICE = torch.device("cuda" if torch.cuda.is_available() else "cpu")
AMP = torch.cuda.is_available()
print("Device:", DEVICE)


# =============================
# Reproducibility
# =============================
def seed_everything(seed: int):
    random.seed(seed)
    np.random.seed(seed)
    torch.manual_seed(seed)
    torch.cuda.manual_seed_all(seed)
    torch.backends.cudnn.deterministic = True
    torch.backends.cudnn.benchmark = False

seed_everything(cfg.seed)


# =============================
# Dataset scan (folder = class)
# =============================
IMG_EXTS = {".jpg",".jpeg",".png",".bmp",".tif",".tiff",".webp"}

def list_images(root: str) -> pd.DataFrame:
    root = Path(root)
    rows = []
    for p in root.rglob("*"):
        if p.is_file() and p.suffix.lower() in IMG_EXTS:
            rows.append((str(p), p.parent.name))
    return pd.DataFrame(rows, columns=["path","label_name"])

df = list_images(cfg.DATA_DIR)
assert len(df) > 0, "No images found. Check cfg.DATA_DIR."
label_names = sorted(df["label_name"].unique().tolist())
label_to_idx = {n:i for i,n in enumerate(label_names)}
df["label"] = df["label_name"].map(label_to_idx).astype(int)
num_classes = len(label_names)

print("Total images:", len(df))
print("Num classes:", num_classes)


# =============================
# Split: 90:10 test, then 10% val from train
# =============================
def make_split(df_all: pd.DataFrame, train_pct: int, test_pct: int, seed: int):
    test_ratio = test_pct / 100.0
    train_all, test_df = train_test_split(
        df_all, test_size=test_ratio, random_state=seed, stratify=df_all["label"]
    )

    # from train_all, take val_in_train for val
    val_ratio = cfg.val_in_train
    val_df, train_df = train_test_split(
        train_all, test_size=(1 - val_ratio), random_state=seed, stratify=train_all["label"]
    )
    return train_df.reset_index(drop=True), val_df.reset_index(drop=True), test_df.reset_index(drop=True)

train_df, val_df, test_df = make_split(df, cfg.train_pct, cfg.test_pct, cfg.seed)
print("Split sizes:", len(train_df), len(val_df), len(test_df))

manifest = {
    "seed": cfg.seed,
    "split": f"{cfg.train_pct}:{cfg.test_pct}",
    "train_paths": train_df["path"].tolist(),
    "val_paths": val_df["path"].tolist(),
    "test_paths": test_df["path"].tolist(),
    "label_to_idx": label_to_idx,
    "idx_to_label": {v:k for k,v in label_to_idx.items()}
}
with open(os.path.join(cfg.WORK_DIR, "split_manifest.json"), "w") as f:
    json.dump(manifest, f)


# =============================
# Transforms
# =============================
normalize = transforms.Normalize(mean=(0.485,0.456,0.406), std=(0.229,0.224,0.225))

train_tf = transforms.Compose([
    transforms.RandomResizedCrop(cfg.img_size, scale=(0.7, 1.0)),
    transforms.RandomHorizontalFlip(p=0.5),
    transforms.ToTensor(),
    normalize,
])

eval_tf = transforms.Compose([
    transforms.Resize(int(cfg.img_size*1.15)),
    transforms.CenterCrop(cfg.img_size),
    transforms.ToTensor(),
    normalize,
])


# =============================
# Torch dataset/loader
# =============================
class ImgDS(Dataset):
    def __init__(self, df: pd.DataFrame, transform):
        self.paths = df["path"].tolist()
        self.labels = df["label"].tolist()
        self.transform = transform
    def __len__(self): return len(self.paths)
    def __getitem__(self, i):
        p = self.paths[i]
        y = self.labels[i]
        img = Image.open(p).convert("RGB")
        x = self.transform(img)
        return x, y, p

train_loader = DataLoader(ImgDS(train_df, train_tf), batch_size=cfg.batch_size, shuffle=True,
                          num_workers=cfg.num_workers, pin_memory=True)
val_loader   = DataLoader(ImgDS(val_df, eval_tf), batch_size=cfg.batch_size, shuffle=False,
                          num_workers=cfg.num_workers, pin_memory=True)
test_loader  = DataLoader(ImgDS(test_df, eval_tf), batch_size=cfg.batch_size, shuffle=False,
                          num_workers=cfg.num_workers, pin_memory=True)


# =============================
# Models (5) — run once each
# =============================
def build_model(model_key: str, n_classes: int) -> nn.Module:
    if model_key == "googlenet":
        m = googlenet(weights=GoogLeNet_Weights.IMAGENET1K_V1)
        m.fc = nn.Linear(m.fc.in_features, n_classes)
        return m
    else:
        return timm.create_model(model_key, pretrained=True, num_classes=n_classes)

MODEL_LIST = [
    ("googlenet", "GoogleNet"),
    ("mobilenetv3_large_100", "MobileNetV3-Large"),
    ("resnet50", "ResNet-50"),
    ("resnet101", "ResNet-101"),
    ("vit_base_patch16_224", "ViT-B/16"),
]


# =============================
# Metrics / plots
# =============================
def per_class_accuracy(y_true, y_pred, n_classes: int) -> Dict[str,float]:
    out = {}
    for c in range(n_classes):
        mask = (y_true == c)
        out[str(c)] = float((y_pred[mask] == y_true[mask]).mean()) if mask.sum() else float("nan")
    return out

def compute_metrics(y_true, y_pred, y_prob, n_classes: int) -> Dict:
    acc = float(accuracy_score(y_true, y_pred))
    prec, rec, f1, _ = precision_recall_fscore_support(
        y_true, y_pred, average="macro", zero_division=0
    )
    y_true_bin = label_binarize(y_true, classes=list(range(n_classes)))
    auc = None
    try:
        auc = float(roc_auc_score(y_true_bin, y_prob, average="macro", multi_class="ovr"))
    except Exception:
        auc = None

    return {
        "accuracy": acc,
        "precision_macro": float(prec),
        "recall_macro": float(rec),
        "f1_macro": float(f1),
        "roc_auc_ovr_macro": auc,
        "per_class_accuracy": per_class_accuracy(y_true, y_pred, n_classes),
    }

def plot_learning(history: Dict, title: str, save_path: str):
    plt.figure(figsize=(8,5))
    plt.plot(history["train_loss"], label="train_loss")
    plt.plot(history["val_loss"], label="val_loss")
    plt.plot(history["train_acc"], label="train_acc")
    plt.plot(history["val_acc"], label="val_acc")
    plt.title(title)
    plt.xlabel("epoch")
    plt.legend()
    plt.tight_layout()
    plt.savefig(save_path, dpi=200)
    plt.show()

def plot_confmat(cm, class_names, title, save_path):
    plt.figure(figsize=(10,8))
    plt.imshow(cm, interpolation="nearest")
    plt.title(title)
    plt.colorbar()
    ticks = np.arange(len(class_names))
    plt.xticks(ticks, class_names, rotation=90)
    plt.yticks(ticks, class_names)
    plt.xlabel("Predicted")
    plt.ylabel("True")
    plt.tight_layout()
    plt.savefig(save_path, dpi=200)
    plt.show()

def plot_roc_curves(y_true, y_prob, n_classes: int, title: str, save_path: str, max_classes=10):
    y_true_bin = label_binarize(y_true, classes=list(range(n_classes)))
    plt.figure(figsize=(8,6))
    plotted = 0
    for c in range(n_classes):
        if y_true_bin[:, c].sum() == 0:
            continue
        fpr, tpr, _ = roc_curve(y_true_bin[:, c], y_prob[:, c])
        plt.plot(fpr, tpr, label=f"{c}")
        plotted += 1
        if plotted >= max_classes:
            break
    plt.plot([0,1],[0,1], linestyle="--")
    plt.title(f"{title} (up to {max_classes} classes)")
    plt.xlabel("FPR")
    plt.ylabel("TPR")
    plt.legend(fontsize=8)
    plt.tight_layout()
    plt.savefig(save_path, dpi=200)
    plt.show()

def top_confused_pairs(cm: np.ndarray, class_names: List[str], topk: int = 12):
    cm2 = cm.copy()
    np.fill_diagonal(cm2, 0)
    pairs = []
    for i in range(cm2.shape[0]):
        for j in range(cm2.shape[1]):
            if cm2[i,j] > 0:
                pairs.append((cm2[i,j], i, j))
    pairs.sort(reverse=True)
    return [{"true": class_names[i], "pred": class_names[j], "count": int(cnt)} for cnt,i,j in pairs[:topk]]

@torch.no_grad()
def estimate_gflops(model: nn.Module, img_size: int = 224) -> float:
    model.eval()
    x = torch.randn(1, 3, img_size, img_size, device=DEVICE)
    flops = FlopCountAnalysis(model.to(DEVICE), x).total()
    return float(flops) / 1e9


# =============================
# Train/Eval loops
# =============================
def run_one_epoch(model, loader, optimizer=None):
    is_train = optimizer is not None
    model.train() if is_train else model.eval()

    total_loss, correct, total = 0.0, 0, 0
    scaler = torch.cuda.amp.GradScaler(enabled=AMP)

    for x, y, _ in loader:
        x = x.to(DEVICE, non_blocking=True)
        y = y.to(DEVICE, non_blocking=True)

        if is_train:
            optimizer.zero_grad(set_to_none=True)

        with torch.cuda.amp.autocast(enabled=AMP):
            logits = model(x)
            loss = F.cross_entropy(logits, y)

        if is_train:
            scaler.scale(loss).backward()
            scaler.step(optimizer)
            scaler.update()

        total_loss += loss.item() * x.size(0)
        pred = logits.argmax(dim=1)
        correct += (pred == y).sum().item()
        total += x.size(0)

    return total_loss/total, correct/total

@torch.no_grad()
def predict_proba(model, loader):
    model.eval()
    all_y, all_pred, all_prob = [], [], []
    t0 = time.time()
    for x, y, _ in loader:
        x = x.to(DEVICE, non_blocking=True)
        logits = model(x)
        prob = F.softmax(logits, dim=1).cpu().numpy()
        pred = logits.argmax(dim=1).cpu().numpy()
        all_prob.append(prob)
        all_pred.append(pred)
        all_y.append(y.numpy())
    infer_time = time.time() - t0
    return np.concatenate(all_y), np.concatenate(all_pred), np.concatenate(all_prob), infer_time


# =============================
# Run all 5 models once
# =============================
results = []

for model_key, model_name in MODEL_LIST:
    print("\n" + "="*80)
    print("Training:", model_name)

    out_dir = os.path.join(cfg.WORK_DIR, model_key)
    os.makedirs(out_dir, exist_ok=True)

    model = build_model(model_key, num_classes).to(DEVICE)
    optimizer = torch.optim.AdamW(model.parameters(), lr=cfg.lr, weight_decay=cfg.weight_decay)
    scheduler = torch.optim.lr_scheduler.CosineAnnealingLR(optimizer, T_max=cfg.epochs)

    # GFLOPs at fixed input size
    try:
        gflops = estimate_gflops(model, cfg.img_size)
    except Exception:
        gflops = None

    history = {"train_loss":[], "val_loss":[], "train_acc":[], "val_acc":[]}

    best_val_acc = -1.0
    best_state = None

    t_train0 = time.time()

    # Force exactly 50 epochs
    for epoch in range(cfg.epochs):
        tr_loss, tr_acc = run_one_epoch(model, train_loader, optimizer=optimizer)
        va_loss, va_acc = run_one_epoch(model, val_loader, optimizer=None)
        scheduler.step()

        history["train_loss"].append(tr_loss)
        history["val_loss"].append(va_loss)
        history["train_acc"].append(tr_acc)
        history["val_acc"].append(va_acc)

        if va_acc > best_val_acc:
            best_val_acc = va_acc
            best_state = {k:v.detach().cpu() for k,v in model.state_dict().items()}

        print(f"epoch {epoch+1:02d}/{cfg.epochs} | tr_acc {tr_acc:.4f} va_acc {va_acc:.4f} | tr_loss {tr_loss:.4f} va_loss {va_loss:.4f}")

    train_time = time.time() - t_train0

    # load best val checkpoint
    if best_state is not None:
        model.load_state_dict(best_state)

    # plots
    plot_learning(history, f"{model_name} | split {cfg.train_pct}:{cfg.test_pct}", os.path.join(out_dir, "learning_curves.png"))

    # test eval
    y_true, y_pred, y_prob, test_infer_time = predict_proba(model, test_loader)
    metrics = compute_metrics(y_true, y_pred, y_prob, num_classes)

    cm = confusion_matrix(y_true, y_pred)
    plot_confmat(cm, label_names, f"Confusion Matrix | {model_name}", os.path.join(out_dir, "confusion_matrix.png"))
    plot_roc_curves(y_true, y_prob, num_classes, f"ROC Curves | {model_name}", os.path.join(out_dir, "roc_curves.png"))

    confused = top_confused_pairs(cm, label_names, topk=12)

    row = {
        "split": f"{cfg.train_pct}:{cfg.test_pct}",
        "model_key": model_key,
        "model_name": model_name,
        "epochs": cfg.epochs,
        "best_val_acc": float(best_val_acc),

        "test_accuracy": metrics["accuracy"],
        "precision_macro": metrics["precision_macro"],
        "recall_macro": metrics["recall_macro"],
        "f1_macro": metrics["f1_macro"],
        "roc_auc_ovr_macro": metrics["roc_auc_ovr_macro"],

        "train_time_sec": float(train_time),
        "test_infer_time_sec": float(test_infer_time),
        "gflops_per_inference_1x3xHxW": gflops,

        "failure_modes_top_confusions": confused
    }

    # Save artifacts
    with open(os.path.join(out_dir, "history.json"), "w") as f:
        json.dump(history, f, indent=2)
    with open(os.path.join(out_dir, "metrics.json"), "w") as f:
        json.dump(metrics, f, indent=2)
    with open(os.path.join(out_dir, "result_summary.json"), "w") as f:
        json.dump(row, f, indent=2)

    results.append(row)

# -----------------------------
# Final comparison tables
# -----------------------------
df_res = pd.DataFrame(results)

csv_path = os.path.join(cfg.WORK_DIR, "task2_results_90_10_all5.csv")
json_path = os.path.join(cfg.WORK_DIR, "task2_results_90_10_all5.json")
df_res.to_csv(csv_path, index=False)
with open(json_path, "w") as f:
    json.dump(results, f, indent=2)

print("\n" + "="*80)
print("DONE. Saved:")
print(csv_path)
print(json_path)

print("\nComparison Table (sorted by TEST accuracy):")
display(df_res[[
    "model_name","test_accuracy","roc_auc_ovr_macro","f1_macro",
    "train_time_sec","test_infer_time_sec","gflops_per_inference_1x3xHxW"
]].sort_values("test_accuracy", ascending=False))

print("\nOnly TEST Accuracy (quick view):")
for _, r in df_res.sort_values("test_accuracy", ascending=False).iterrows():
    print(f"{r['model_name']}: {r['test_accuracy']:.4f}")
```
